# 2: Pandas, Scatter Plots
Most of the plotting functionality in pandas is contained within the DataFrame.plot() method. When we call this method, we specify the data we want plotted as well as the type of plot. 
* We use the kind parameter to specify the type of plot we want. 
* We use x and y to specify the data we want on each axis.
```recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')```

````python
recent_grads.plot(x='Sample_size', 
                  y='Median', 
                  kind='scatter', 
                  title='Sample_size vs. Median',
                  figsize=(4, 2))
````

````python
#To tweak plot, assign the returned axes object to a variable
ax = recent_grads.plot(x='Sample_size', 
                  y='Unemployment_rate', 
                  kind='scatter', 
                  figsize=(4, 2))
ax.set_title('Sample_size vs. Unemployment_rate')
````

````python
recent_grads.plot(x='Full_time', 
                  y='Median', 
                  kind='scatter', 
                  title='Full_time vs. Median',
                  figsize=(4, 2))
recent_grads.plot(x='ShareWomen', 
                  y='Unemployment_rate', 
                  kind='scatter', 
                  title='ShareWomen vs. Unemployment_rate',
                  figsize=(4, 2))
recent_grads.plot(x='Men', 
                  y='Median', 
                  kind='scatter', 
                  title='Men vs. Median',
                  figsize=(4, 2))
recent_grads.plot(x='Women', 
                  y='Median', 
                  kind='scatter', 
                  title='Women vs. Median',
                  figsize=(4, 2))
````

# 4: Pandas, Scatter Matrix Plot
A scatter matrix plot combines both scatter plots and histograms into one grid of plots and allows us to explore potential relationships and distributions simultaneously. A scatter matrix plot consists of n by n plots on a grid, where n is the number of columns, the plots on the diagonal are histograms, and the non-diagonal plots are scatter plots.

````python
from pandas.tools.plotting import scatter_matrix

columns2 = ['Sample_size', 'Median']
columns3 = ['Sample_size', 'Median', 'Unemployment_rate']
scatter_matrix(recent_grads[columns2])
````

# 5: Pandas, Bar Plots

````python
# using matplotlib to plot bar chart needs specify the locations, labels, lengths and widths of the bar.
# Using pandas plotting is much easier
recent_grads[:5]['Women'].plot(kind='bar')
````

````python
recent_grads[:5]['Women'].plot(kind='bar', x='Major', y='Women')
````

````python
# Note that plot() function can't customize anything about the chart.
# It uses default x-tick label (from 0 - n)
# To get more flexibility, use bar() function
recent_grads[:5].plot.bar(x='Major', y="Women")
````
